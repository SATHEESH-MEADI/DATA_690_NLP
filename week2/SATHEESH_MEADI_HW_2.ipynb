```python
import urllib.request
from bs4 import BeautifulSoup

#Connect to the website
url = 'https://pythonprogramming.net/parsememcparseface/'
response = urllib.request.urlopen(url)

# Read the HTML content
html = response.read()

#Parse the HTML with BeautifulSoup
soup = BeautifulSoup(html, 'lxml')

# Print the parsed content (you can modify this to extract specific data)
print(soup.prettify())
```

```python
# Print the page title
print("Page Title:", soup.title.string)
```

```python

# Get all attributes and values from the first <a> tag
first_link = soup.find('a')
if first_link:
    print("First Link Attributes:")
    for attribute, value in first_link.attrs.items():
        print(f"{attribute}: {value}")

    # Get specific values from the first link (for example, 'href' and 'text')
    specific_href = first_link.get('href')
    specific_text = first_link.get_text()
    print("Specific Href:", specific_href)
    print("Specific Text:", specific_text)
```

```python
# Find all paragraphs and iterate through them
paragraphs = soup.find_all('p')
for paragraph in paragraphs:
    print("Paragraph Text:", str(paragraph.text))
```

```python
#  Grab all the links
links = soup.find_all('a')
for link in links:
    print("Link:", link.get('href'))
```

```python
# Grab the text from all links
for link in links:
    print("Link Text:", link.get_text())
s
```
